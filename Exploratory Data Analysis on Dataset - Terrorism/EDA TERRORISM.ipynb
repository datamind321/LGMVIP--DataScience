## EDA
Countplot: Number Of Terrorist Activities Each Year

```python
plt.subplots(figsize=(15,6))
sns.countplot('iyear',data=df,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

**Histplot: Number Of Terrorist Activities Each Year**

```python
df.iyear.plot(kind = 'hist', color = 'g', bins=range(1970, 2018), figsize = (16,7),grid=True)
plt.xticks(range(1970, 2018), rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.title("Number of Attacks By Year", fontsize=16)
plt.show()
```

**Countplot: Number Of Terrorist Activities Each Month**

```python
plt.subplots(figsize=(15,6))
sns.countplot('imonth',data=df,edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

**Histplot: Number Of Terrorist Activities Each Month**

```python
df.imonth.plot(kind = 'hist', color = 'g', bins=range(1, 13), figsize = (16,7),grid=True)
plt.xticks(range(1, 13), rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.title("Number of Attacks By Year", fontsize=16)
plt.show()
```

**Countplot: Number Of Terrorist Activities Each Day**

```python
plt.subplots(figsize=(10,5))
sns.countplot('iday',data=df,edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities in each country')
plt.show()
```

**Histplot: Number Of Terrorist Activities Each Day**

```python
df.iday.plot(kind = 'hist', color = 'g', bins=range(1, 31), figsize = (16,7),grid=True)
plt.xticks(range(1, 31), rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Years", fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.title("Number of Attacks By Year", fontsize=16)
plt.show()
```

**Countplot of Total attacks in each region**

```python
region_attacks = df.region.value_counts().to_frame().reset_index()
region_attacks.columns = ['region', 'Total Attacks']
plt.subplots(figsize=(15,6))
sns.barplot(x=region_attacks.region, y=region_attacks['Total Attacks'])
plt.xticks(rotation=40)
plt.title('Number of Total Attacks in Each Region')
plt.show()
```

**Countplot of region and country**

```python
plt.subplots(figsize=(10,5))
sns.countplot('region',data=df,edgecolor=sns.color_palette('dark',16),hue='imonth')
plt.xticks(rotation=30)
plt.title('Number Of Terrorist Activities in each country')
plt.show()
```

```python
region_attacks = df.region.value_counts().to_frame().reset_index()
region_attacks.columns = ['region', 'Total Attacks']
plt.subplots(figsize=(10,5))
sns.histplot(x=region_attacks.region, y=region_attacks['Total Attacks'] ,edgecolor=sns.color_palette('dark',10),kde=True)
plt.xticks(rotation=50)
plt.title('Number of Total Attacks in Each Region')
plt.show()
```

**Attacks**

```python
plt.subplots(figsize=(13,6))
sns.countplot(y='attacktype1_txt',data=df)
plt.title('Favorite Attack')
plt.show()
```

```python
plt.subplots(figsize=(13,45))
sns.countplot(y='country_txt',data=df)
plt.title('Number of attacks per country')
plt.show()
```

```python
plt.subplots(figsize=(10,5))
sns.countplot('imonth',data=df,hue='region',dodge=True)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities in each country')
plt.show()
```
